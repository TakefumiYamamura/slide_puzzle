```python
%matplotlib inline
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

def read_lines(fname):
    ar = []
    for l in open(fname, 'r'):
        ar.append(0.01 if float(l)==0 else float(l))
    return ar
plt.style.use('ggplot')

def plot(y1, y1lab, title, ylabel="Duration in seconds"):
    x = np.arange(0, len(y1), 1)
    plt.title(title)
    plt.xlabel("Number of the problem")
    plt.ylabel(ylabel)
    plt.plot(x, y1, '-', label=y1lab)
    plt.legend(loc = 'upper left')

def time_plt(y1, y1lab, y2, y2lab, title, ylabel="Duration in seconds"):
    com_len = min(len(y1), len(y2))
    x = np.arange(0, com_len, 1)
    plt.title(title)
    plt.xlabel("Number of the problem")
    plt.ylabel(ylabel)
    plt.plot(x, y1[0:com_len], '-', label=y1lab)
    plt.plot(x, y2[0:com_len], '--', label=y2lab)
    plt.legend(loc = 'upper left')
    
def rel_plt(y1, y2, leg):
    com_len = min(len(y1), len(y2))
    x = np.arange(0, com_len, 1)
    baseline = [1] * com_len
    ratio = []
    for i in range(0, com_len):
        ratio.append(y1[i] / y2[i])
    plt.title("Relative performance")
    plt.xlabel("Number of the problem")
    plt.ylabel("Duration ratio")
    plt.plot(x, ratio, label=leg)
    plt.plot(x, baseline, label="y = 1")
    plt.legend(loc = 'upper left')
    
def p_summery(y1, y2):
    print("len1=%s, len2=%s" % (len(y1), len(y2)))
    com_len = min(len(y1), len(y2))
    s1 = sum(y1[0:com_len])
    s2 = sum(y2[0:com_len])
    print("sum(prob 0-%s): s1=%s, s2=%s" %(com_len-1, s1, s2))
    print("relative performance over problems: %s" % (s1/s2))
    
def bar(data, title):
    x = np.arange(0, 32, 1)
    plt.bar(x, data)
    plt.xlim([0,32])
    plt.title(title)
    plt.xlabel("NPS/LPS(round down)")
    plt.ylabel("block count")
```

```python
burns = read_lines('aws_15_burns')
cpu = read_lines('aws_15_cpu')    
time_plt(burns, 'Solver B', cpu, 'Solver C', "Experiment: Solver B,C")
```

```python
rel_plt(burns, cpu, "Solver B / Solver C")
```

```python
p_summery(burns, cpu)
```
